```python
def numprimo(num):
    # Verificar que el número sea mayor que 1
    while num <= 1:
        print("Error en el Número")
        num = int(input("Ingrese un número entero mayor que 1: "))

    # Verificar si es primo
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False

    return True

# Ingresar un número entero mayor que 1
num = int(input("Ingrese un número entero mayor que 1: "))

#imprimir el resultado
resultado = numprimo(num)
print(resultado)
```

```python
#El promedio

num = list(range(151))
promedio = sum(num) / len(num)

print("Promedio:", promedio)
```

```python
#La varianza
num = list(range(151))

media = sum(num) / len(num)

sumacuadrados = sum((x - media) ** 2 for x in num)

varianza = sumacuadrados / len(num)

print("La Varianza es:", varianza)
```

```python
#La desviación estándar
num = list(range(151))

media = sum(num) / len(num)

sumacuadrados = sum((x - media) ** 2 for x in num)

varianza = sumacuadrados / len(num)

desviacionestandar = varianza**.5

print("La Desviación Estándar es:", desviacionestandar)
```

```python
#Visualice los datos en un histograma.
import matplotlib.pyplot as plt

num = list(range(151))

plt.hist(num, edgecolor='black')

plt.title('Histograma')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.show()
```

```python
#La suma de los números pares e impares.
num= list(range(151))

pares = sum(x for x in num if x % 2 == 0)
impares = sum(x for x in num if x % 2 != 0)

print("Suma de números pares: ",pares)
print("Suma de números impares: ", impares)
```

```python
#Los múltiplos de 2, 3 y 5.
numeros = list(range(151))

mul_2 = [x for x in numeros if x % 2 == 0]
mul_3 = [x for x in numeros if x % 3 == 0]
mul_5 = [x for x in numeros if x % 5 == 0]

print("Múltiplos de 2:", mul_2)
print("Múltiplos de 3:", mul_3)
print("Múltiplos de 5:", mul_5)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#valores de C > 0
def y(x, C):
    return (x - 1) + C * np.exp(-x)

x_vals = np.linspace(-3, 3, 100)

plt.plot(x_vals, y(x_vals, 2), label='C > 0', color='blue')


plt.title('Familia de Curvas para la Ecuación Diferencial')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```

```python
#valores de C = 0
def y(x, C):
    return (x - 1) + C * np.exp(-x)

x_vals = np.linspace(-3, 3, 100)

plt.plot(x_vals, y(x_vals, 0), label='C = 0', color='green')

plt.title('Familia de Curvas para la Ecuación Diferencial')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```

```python
#valores de C < 0
def y(x, C):
    return (x - 1) + C * np.exp(-x)

x_vals = np.linspace(-3, 3, 100)

plt.plot(x_vals, y(x_vals, -2), label='C < 0', color='yellow')

plt.title('Familia de Curvas para la Ecuación Diferencial')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```

```python
# Grafique la función con línea continua para un rango de -10 a 10, con más de 1000 muestras

def f(x):
    return (3*x + 7) / (x**2 - 4)

x = np.linspace(-10, 10, 1000)
y = f(x)

plt.plot(x, y, label='(3x + 7) / (x^2 - 4)')

plt.title('Gráfica de la función (3x + 7) / (x^2 - 4)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

plt.show()
```

```python
#Investigue en la literatura cómo graficar las asíntotas horizontales y verticales. Muéstrelas en el gráfico con líneas punteadas.
def f(x):
    return (3*x + 7) / (x**2 - 4)

x_vals = np.linspace(-10, 10, 1000)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label='(3x + 7) / (x^2 - 4)')

plt.axhline(0, color='red', linestyle='--', linewidth=1, label='Asíntota Horizontal')
plt.axvline(-2, color='black', linestyle='--', linewidth=1, label='Asíntota Vertical')
plt.axvline(2, color='black', linestyle='--', linewidth=1)

plt.title('Gráfica de la función con Asíntotas')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

plt.show()
```

```python
# Establezca límites en el eje "y" para evitar que la gráfica muestre valores muy altos o muy bajos cerca de las asíntotas verticales.
def f(x):
    return (3*x + 7) / (x**2 - 4)

x_vals = np.linspace(-10, 10, 1000)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label='(3x + 7) / (x^2 - 4)')

plt.axhline(0, color='black', linestyle='--', linewidth=1, label='Asíntota Horizontal')
plt.axvline(-2, color='red', linestyle='--', linewidth=1, label='Asíntota Vertical')
plt.axvline(2, color='red', linestyle='--', linewidth=1)

plt.ylim(-10, 10)

plt.title('Gráfica de la función con Asíntotas y Límites en el eje y')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

plt.show()
```
