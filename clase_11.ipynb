```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,6*np.pi,100)
coseno=np.cos(x)
seno=np.sin(x)

plt.plot(x,coseno)
plt.plot(x,seno)

plt.title("Trigonometrica")
plt.xlabel("Abcisas")
plt.ylabel("Seno y coseno")

plt.plot(x, y)

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```
