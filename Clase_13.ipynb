```python
SL = df[df['Country']=='Sierra Leone']
```

```python
SL = SL.drop('Country',axis=1)
```

```python
SL.isnull().sum()
```

```python
mean_values=SL['No. of confirmed deaths'].mean()
mean_values
```

```python
SL.fillna(0.0,inplace=True)
```

```python
SL.isnull().sum()
```

```python
SL['No. of confirmed deaths'].mean() #casos totales
```

```python
SL['No. of confirmed cases'].mean() #promedio de casos
```

```python
Dates=SL.index[0:-1:10]
NCC=SL['No. of confirmed cases'][0:-1:10]
NCD=SL['No. of confirmed deaths'][0:-1:10]
```

```python
#Graficacion de los datos de muertes y casos confirmados
plt.plot (Dates,NCC,'*--g',ms=7,label='Infectados')
plt.bar (Dates,NCC,color='green',alpha=0.3)
plt.plot (Dates,NCD,'*--b',ms=8,label='Decesos')
plt.bar (Dates,NCD,color='blue',alpha=0.3)
plt.grid()
plt.title('Avance del ebola en Sierra Leona')
plt.ylabel('Casos confirmados')
plt.xlabel('Periodo de tiepo')
plt.xticks(rotation=90,fontsize=8)
plt.legend()
plt.savefig("Avance_del_ebola_en_Sierra_Leona.pdf",dpi=300,format='pdf') #sirve para poder guardar la imagen pfd
plt.show()
```

```python
PIvsD=100*SL['No. of confirmed deaths']/SL['No. of confirmed cases']
SL['PIvsD']=PIvsD
```

```python
SL.head()
```

```python
recuperados=((100-PIvsD)/100)*SL['No. of confirmed cases']
SL['recuperados']=recuperados
```

```python
SL.head()
```

```python
SL['PR']=(100-PIvsD)
```

```python
PIvsD_mean=SL['PIvsD'].mean()
PR_mean=SL['PR'].mean()
porcentajes=[PIvsD_mean,PR_mean]
etiquetas=['PIvsD','PR']
```

```python
plt.pie(porcentajes,labels=etiquetas,autopct='%0.1f %%', colors=['pink','magenta'])
plt.axis("equal")
plt.title("Porcentaje de Recuperados contra Decesos")
plt.savefig("Porcentaje_de_Recuperados_contra_Decesos.pdf",dpi=300,format='pdf')
plt.show()
```
